```python
import torch
import forward
import backward
import polymerDiffusion as model
import matplotlib.pyplot as plt
```

```python
tmin=5
tmax=15
mu=1
lam=0.96
D0=0.1
nmin=3
nmax=20
smax=0.5
alpha=1
batch_size=200
int_Ns=20
int_Nt=100
epoch=2000
w_f=1
w_int=1.5
w_u0=1.5
```

```python
forward_model = forward.PINNs_v1(mu,lam,D0,nmin,nmax,smax,tmin,tmax,alpha,batch_size,int_Ns,int_Nt)
backward_model = backward.PINNs_v1(mu,lam,D0,nmin,nmax,smax,tmin,tmax,alpha,batch_size,int_Ns,int_Nt)
```

```python
forward_loss_f,forward_loss_int,forward_loss_u0 = model.Polymer_Diffusion.train_PINNs_v1(
    model=forward_model,
    epoch=epoch,
    w_f=w_f,
    w_int=w_int,
    w_u0=w_u0
)
```

```python
backward_loss_f,backward_loss_int,_ = model.Polymer_Diffusion.train_PINNs_v1(
    model=backward_model,
    epoch=epoch,
    w_f=w_f,
    w_int=w_int,
    w_u0=w_u0
)
```

```python
s = torch.arange(0,0.5,0.005)
t = torch.arange(5,15,0.1)
s,t = torch.meshgrid(s,t)
```

```python
s_ = s.reshape([s.shape[0]*s.shape[1],1])
t_ = t.reshape([s.shape[0]*s.shape[1],1])
ans_forward = forward_model.mlp(torch.cat([s_,t_],dim=1)).reshape([s.shape[0],s.shape[1],21]).detach().numpy()
ans_backward = backward_model.mlp(torch.cat([s_,t_],dim=1)).reshape([s.shape[0],s.shape[1],21]).detach().numpy()
```

```python
for k in range(21):
    plt.xlabel('t')
    plt.ylabel('s')
    plt.contourf(t,s,ans_forward[:,:,k],levels=100, alpha=1, cmap='jet')
    plt.colorbar()
    plt.savefig('../../figure/forward_'+str(k)+'n.png', dpi=1000, bbox_inches='tight')
    plt.show()
    plt.xlabel('t')
    plt.ylabel('s')
    plt.contourf(t,s,ans_backward[:,:,k],levels=100, alpha=1, cmap='jet')
    plt.colorbar()
    # plt.savefig('../../figure/backward_'+str(k)+'n.png', dpi=1000, bbox_inches='tight')
    plt.show()
```

```python
plt.xlabel('t')
plt.ylabel('s')
plt.contourf(t.numpy(),s.numpy(),ans_forward[:,:,20],levels=100, alpha=1, cmap='jet')
plt.colorbar()
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(10,10))
plt.xlabel('s')
plt.ylabel('u(n,s,t)')
mm = 0
nn = 2
int_ = np.zeros(nmax+1)
for kk in range(int(21/nn)):
    for i in range(nmax+1):
        int_[i] = np.sum(ans_forward[:,mm,i])*0.5/100
    ss = s[:,mm]
    yy = ans_forward[:,mm,kk*nn]
    plt.plot(ss.detach().numpy(),yy,label='n='+str(kk*nn))
plt.legend()
plt.show()
print(int_)
```
